# DATA PRE-PROCESSING

```python
# --- Isolate the Feature Matrix (X) ---
# We create a new DataFrame 'X' that contains all the columns from our original DataFrame 'df'
# EXCEPT for the 'Class' column.
# The .drop() method is used for this.
# 'Class' is the name of the column we want to remove.
# 'axis=1' is a crucial argument. It tells pandas that we are dropping a COLUMN.
# If we used axis=0, pandas would try to drop a row with the index label 'Class', which doesn't exist.
X = df.drop('Class', axis=1)

# --- Isolate the Target Vector (y) ---
# We create a pandas Series 'y' that contains ONLY the 'Class' column.
# This is our target variable, the value our model will learn to predict.
# Standard pandas syntax for selecting a single column is square bracket notation.
y = df['Class']

# --- Verify the Separation ---
# It's always a good practice to check the results of our operations.
# Let's print the shapes and the first few rows of our new variables.

print("Shape of the Feature Matrix (X):", X.shape)
print("Shape of the Target Vector (y):", y.shape)

print("\nFirst 5 rows of the Feature Matrix (X):")
print(X.head())

print("\nFirst 5 rows of the Target Vector (y):")
print(y.head())
```

```python
# Import the RobustScaler from scikit-learn's preprocessing module.
from sklearn.preprocessing import RobustScaler

# --- Step 1: Instantiate the Scaler ---
# Create an instance of the RobustScaler.
# We will use this object to scale our data.
scaler = RobustScaler()

# --- Step 2: Select and Scale the Columns ---
# We will scale the 'Time' and 'Amount' columns from our feature matrix 'X'.
# We create a copy of these columns to avoid a SettingWithCopyWarning and to keep the original X intact for now.
X_scaled = X.copy()
columns_to_scale = ['Time', 'Amount']

# The fit_transform() method is a convenient two-in-one function:
# 1. 'fit': The scaler calculates the median and IQR for the 'Time' and 'Amount' columns.
# 2. 'transform': It then uses these calculated statistics to scale the data.
# The output is a NumPy array with the scaled values.
X_scaled[columns_to_scale] = scaler.fit_transform(X[columns_to_scale])

# --- Step 3: Verify the Transformation ---
# Let's look at the first few rows of our newly scaled data.
# Notice how the values in 'Time' and 'Amount' are now much smaller and centered around zero.
print("First 5 rows of the scaled features:")
print(X_scaled[['Time', 'Amount']].head())

print("\n" + "-"*50 + "\n")

# To truly see the effect, let's compare the statistical summary of the original and scaled 'Amount' column.
print("Original 'Amount' feature statistics:")
print(X['Amount'].describe())

print("\n" + "-"*50 + "\n")

print("Scaled 'Amount' feature statistics:")
print(X_scaled['Amount'].describe())
```

```python
# Import the pandas library if it's not already in the cell (it's good practice for cell-based work).
import pandas as pd

# --- Step 1: Isolate the two feature sets we want to combine ---

# Isolate the scaled 'Time' and 'Amount' columns from the X_scaled DataFrame we created in the last step.
scaled_features = X_scaled[['Time', 'Amount']]

# Isolate the original V1-V28 features from our original feature matrix 'X'.
# We can do this by dropping the 'Time' and 'Amount' columns from 'X'.
v_features = X.drop(['Time', 'Amount'], axis=1)

# --- Step 2: Concatenate the feature sets ---

# Now, we use pd.concat() to join our two DataFrames side-by-side.
# The function takes a list of the DataFrames to be concatenated: [v_features, scaled_features].
# The crucial argument is `axis=1`, which tells pandas to join them horizontally (as new columns).
# Pandas will use the index to align the rows perfectly.
X_processed = pd.concat([v_features, scaled_features], axis=1)

# --- Step 3: Verify the result ---

# Let's inspect the shape and the first few rows of our final processed feature matrix.
print("Shape of the final processed feature matrix (X_processed):", X_processed.shape)

# The head of the DataFrame should show the original 'V' features and the newly scaled 'Time' and 'Amount' columns.
# Note that the order of columns might be different from the original 'X' based on our concatenation order,
# but this has no impact on the model's performance.
print("\nFirst 5 rows of the final processed feature matrix:")
X_processed.head()
```

```python
# Import the train_test_split function again for clarity in this cell.
from sklearn.model_selection import train_test_split

# We perform the split again, but this time we add the 'stratify' parameter.
#
# stratify=y: This is the crucial addition. We pass the target vector 'y' to this parameter.
# Scikit-learn will look at the distribution of values in 'y' (the proportions of 0s and 1s)
# and ensure that the split preserves this exact same proportion in both y_train and y_test.
# This guarantees that our training and testing sets are perfect representations of the overall dataset's imbalance.
X_train, X_test, y_train, y_test = train_test_split(
    X_processed,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y  # <--- THE CRUCIAL ENHANCEMENT
)

# --- Verification Step: Let's prove that stratification worked! ---
# We will calculate the percentage distribution of classes in the original, training, and testing sets.

print("Original dataset class distribution (%):")
print(round(y.value_counts(normalize=True) * 100, 3))

print("\nTraining set class distribution (%):")
print(round(y_train.value_counts(normalize=True) * 100, 3))

print("\nTesting set class distribution (%):")
print(round(y_test.value_counts(normalize=True) * 100, 3))
```

```python
# SMOTE stands for Synthetic Minority Over-sampling Technique.
# It is a powerful algorithm used to generate synthetic samples for the minority class.
# We import the SMOTE class from the `over_sampling` module within the `imblearn` library.
# This makes the tool available for us to use in the subsequent steps.
from imblearn.over_sampling import SMOTE
```

```python
# We are now creating an instance of the SMOTE class.
# The variable 'smote' will hold our configured SMOTE object.
#
# random_state=42: This is the most crucial parameter we set. As explained, SMOTE involves
# random processes. By setting a random_state, we ensure that the synthetic samples
# generated are identical every time we run this code. This makes our entire
# experiment reproducible, which is a cornerstone of good data science practice.
#
# Note: SMOTE has other parameters, like k_neighbors (default is 5), which controls how many
# neighbors are considered when generating a new sample. For now, the default values are
# perfectly suitable for our needs.
smote = SMOTE(random_state=42)

# To confirm the object has been created, we can simply print it.
# This will show us the object in memory along with the parameters we've set.
print(smote)
```

```python
# --- Applying SMOTE to the Training Data ---
# We call the .fit_resample() method on our instantiated smote object.
# This method takes our imbalanced training data (X_train and y_train) as input.
# It performs the SMOTE algorithm and returns two new, balanced datasets.
#
# X_resampled: A new feature matrix containing all original non-fraudulent samples
#              plus the original fraudulent samples AND the newly created synthetic ones.
# y_resampled: A new target vector corresponding to X_resampled, now with a 50/50
#              distribution of 0s and 1s.
#
# This is a crucial step: we ONLY resample the training data.
# The test set (X_test, y_test) remains untouched to serve as a realistic, imbalanced
# evaluation set that mirrors the real world.
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# --- Verifying the Transformation by Checking Shapes ---
# The most immediate way to see the effect of SMOTE is to compare the shapes of our
# datasets before and after resampling.
print("Shape of original training data (X_train):", X_train.shape)
print("Shape of original training labels (y_train):", y_train.shape)
print("-" * 50)
print("Shape of resampled training data (X_resampled):", X_resampled.shape)
print("Shape of resampled training labels (y_resampled):", y_resampled.shape)
```

```python
# --- Numerical Verification ---
# To see the "before and after" effect, we'll first check the class distribution
# in our original, imbalanced training set.
print("Class Distribution Before SMOTE:")
print(y_train.value_counts())

print("\n" + "-"*50 + "\n") # A separator for clarity

# Now, we check the class distribution in our new, resampled training set.
# This is the moment of truth where we confirm if the classes are balanced.
print("Class Distribution After SMOTE:")
print(y_resampled.value_counts())
```

```python
# Import visualization libraries for this cell.
import seaborn as sns
import matplotlib.pyplot as plt

# Create a countplot for the resampled data.
plt.figure(figsize=(8, 6))
sns.countplot(x=y_resampled)
plt.title('Class Distribution After SMOTE')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Frequency')
plt.show()
```

# MODEL TRAINING 

```python
# Import the LogisticRegression class from scikit-learn's linear_model module.
from sklearn.linear_model import LogisticRegression

# --- Step 1: Instantiate the Model ---
# We create an instance of the LogisticRegression model.
# We will store this instance in a variable called `lr_baseline`.
#
# random_state=42: While the default solver for Logistic Regression is deterministic,
# other solvers might have random elements. Setting a random_state is a good habit
# for ensuring our experiments are 100% reproducible across the board.
#
# max_iter=1000: This parameter sets the maximum number of iterations for the solver
# to converge (i.e., to find the best model parameters). The default is 100.
# For some datasets, especially after scaling, the solver might need more iterations
# to find the optimal solution. Setting it to 1000 is a safe-guard against
# "convergence warning" errors.
lr_baseline = LogisticRegression(random_state=42, max_iter=1000)

# --- Step 2: Train the Model ---
# This is the core training step. We use the .fit() method of our model instance.
# .fit() takes two main arguments:
#  1. The feature matrix (X): The "questions" or "evidence".
#  2. The target vector (y): The "answers" or "verdicts".
# The model will analyze X_train and y_train to learn the relationship between the features
# and the class labels.
#
# CRUCIALLY, we are using X_train and y_train - our original, IMBALANCED data.
print("Training the baseline Logistic Regression model on the original, imbalanced data...")
lr_baseline.fit(X_train, y_train)
print("Baseline model trained successfully!")

# The trained model object `lr_baseline` now contains the learned "knowledge".
# We can see the object itself to confirm it has been fitted.
print("\nFitted Model Object:")
print(lr_baseline)
```

```python
# We will use the same LogisticRegression class, as the algorithm type is the same.
# If it's not in your current cell, you can re-import it for clarity.
from sklearn.linear_model import LogisticRegression

# --- Step 1: Instantiate a NEW Model ---\
# We create a second instance of the LogisticRegression model.
# It's critical to create a new object. We'll name it `lr_smote` to signify
# that it will be trained on the SMOTE-balanced data.
# We use the exact same parameters (random_state=42, max_iter=1000) as the baseline
# model. This ensures a fair comparison, where the ONLY difference is the
# training data itself.
lr_smote = LogisticRegression(random_state=42, max_iter=1000)

# --- Step 2: Train the Model on Resampled Data ---\
# This is the key step. We once again use the .fit() method.
# However, this time we pass our balanced training data: X_resampled and y_resampled.
# The model is now learning from a dataset with a perfect 50/50 split of fraud and non-fraud cases.
print("Training the Logistic Regression model on the SMOTE-resampled (balanced) data...")
lr_smote.fit(X_resampled, y_resampled)
print("SMOTE-trained model trained successfully!")

# --- Step 3: Confirm Training ---
# The `lr_smote` object now contains the knowledge learned from the balanced data.
# This knowledge (the internal coefficients) will be different from the `lr_baseline` model.
print("\nFitted SMOTE Model Object:")
print(lr_smote)
```

```python
# Import the RandomForestClassifier from scikit-learn's powerful `ensemble` module.
from sklearn.ensemble import RandomForestClassifier

# --- Step 1: Instantiate the Model ---
# We create an instance of the RandomForestClassifier.
# We'll name it `rf_smote` to be clear it's a Random Forest trained on our SMOTE data.
#
# n_estimators=100: This is a key hyperparameter. It specifies that we want to build
#                   a forest containing 100 individual decision trees.
#
# random_state=42: As with our other models, this is crucial for reproducibility.
#                   The process of creating data samples (bagging) and selecting feature
#                   subsets involves randomness. This ensures the exact same forest is
#                   built every time we run the code.
#
# n_jobs=-1: This is a fantastic performance parameter. It tells scikit-learn to use
#            all available CPU cores on your machine to train the trees in parallel.
#            For a model like Random Forest, this can dramatically reduce training time.
rf_smote = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# --- Step 2: Train the Model ---
# We use the familiar .fit() method, feeding it our best training data:
# the balanced feature matrix `X_resampled` and target vector `y_resampled`.
print("Training the Random Forest Classifier on the SMOTE-resampled (balanced) data...")
rf_smote.fit(X_resampled, y_resampled)
print("Random Forest model trained successfully!")

# --- Step 3: Confirm Training ---
# The `rf_smote` object now contains 100 trained decision trees, ready to vote on new data.
print("\nFitted Random Forest Model Object:")
print(rf_smote)
```

```python
# Import the XGBClassifier from the xgboost library.
# You installed this during your initial environment setup.
from xgboost import XGBClassifier

# --- Step 1: Instantiate the Model ---
# We create an instance of the XGBClassifier.
# We will name it `xgb_smote` to maintain our naming convention.
#
# n_estimators=100: Same as our Random Forest, we'll build an ensemble of 100 trees.
#
# use_label_encoder=False: In recent versions of XGBoost, it's recommended to set this to False
#                          to avoid a deprecation warning. It tells XGBoost not to use an old
#                          scikit-learn label encoding mechanism. Our labels are already 0 and 1,
#                          so this is perfectly fine.
#
# random_state=42: Absolutely crucial. Boosting involves several random elements (like feature
#                  subsampling). Setting the random_state ensures the same sequence of trees
#                  is built every time, making our results reproducible.
#
# eval_metric='logloss': This explicitly sets the evaluation metric to be used for internal
#                        model improvements. 'logloss' is a standard for binary classification
#                        and helps suppress potential warnings.
#
# n_jobs=-1: Just like with Random Forest, this tells XGBoost to use all available CPU
#            cores, which can significantly speed up the training process.
xgb_smote = XGBClassifier(n_estimators=100, use_label_encoder=False, random_state=42, eval_metric='logloss', n_jobs=-1)

# --- Step 2: Train the Model ---
# We fit the XGBoost model on our balanced, SMOTE-resampled data.
# The model will sequentially build 100 trees, each one learning from the errors of the last.
print("Training the XGBoost Classifier on the SMOTE-resampled (balanced) data...")
xgb_smote.fit(X_resampled, y_resampled)
print("XGBoost model trained successfully!")

# --- Step 3: Confirm Training ---
# The `xgb_smote` object now holds the fully trained, sequential ensemble of trees.
print("\nFitted XGBoost Model Object:")
print(xgb_smote)
```

```python
# Import the LGBMClassifier from the lightgbm library.
# You installed this during your initial environment setup.
from lightgbm import LGBMClassifier

# --- Step 1: Instantiate the Model ---
# We create an instance of the LGBMClassifier, following our naming convention.
#
# n_estimators=100: We will build an ensemble of 100 trees, keeping it consistent
#                   with our Random Forest and XGBoost models for a fair comparison.
#
# random_state=42: This is absolutely essential for reproducibility. LightGBM, like
#                  other boosting algorithms, has random elements in its training
#                  process (e.g., feature subsampling). This guarantees that
#                  the same model is built every time.
#
# n_jobs=-1: This performance-boosting parameter tells LightGBM to use all available
#            CPU cores for training, which can dramatically reduce training time.
lgbm_smote = LGBMClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# --- Step 2: Train the Model ---
# We use the familiar .fit() method on our balanced, SMOTE-resampled training data.
# The model will now build its 100 trees using the efficient leaf-wise growth strategy.
print("Training the LightGBM Classifier on the SMOTE-resampled (balanced) data...")
lgbm_smote.fit(X_resampled, y_resampled)
print("LightGBM model trained successfully!")

# --- Step 3: Confirm Training ---
# The `lgbm_smote` object now contains the fully trained LightGBM ensemble.
print("\nFitted LightGBM Model Object:")
print(lgbm_smote)
```

```python
# Import the necessary functions from scikit-learn's metrics module
# precision_recall_curve calculates the points for the P-R curve
# auc is used to calculate the area under any curve, which we'll use for AUPRC
from sklearn.metrics import precision_recall_curve, auc

# Create a dictionary of our FITTED model objects for easy iteration
# This is different from the predictions dictionary; here we need the models themselves
# to access their .predict_proba() method.
models = {
    "Logistic Regression (Baseline)": lr_baseline,
    "Logistic Regression (SMOTE)": lr_smote,
    "Random Forest (SMOTE)": rf_smote,
    "XGBoost (SMOTE)": xgb_smote,
    "LightGBM (SMOTE)": lgbm_smote
}

# --- Dictionary to store the final AUPRC scores ---
auprc_scores = {}

print("--- Calculating AUPRC for each model ---")

# --- Loop through each model to calculate its AUPRC score ---
for model_name, model in models.items():
    # Step 1: Get prediction probabilities for the POSITIVE class (class '1')
    # The .predict_proba(X_test) method returns an array with two columns:
    # Column 0: Probability of the class being '0' (Not Fraud)
    # Column 1: Probability of the class being '1' (Fraud)
    # We are interested in the model's confidence in predicting fraud, so we take the second column [:, 1].
    y_proba = model.predict_proba(X_test)[:, 1]
    
    # Step 2: Calculate the points for the Precision-Recall curve
    # This function takes the true labels (y_test) and the positive class probabilities (y_proba)
    # It returns arrays of precision and recall values for each unique threshold.
    precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
    
    # Step 3: Calculate the Area Under the Curve (AUPRC)
    # The auc() function takes the x-coordinates (recall) and y-coordinates (precision)
    # and computes the area underneath the resulting curve.
    auprc_score = auc(recall, precision)
    
    # Store the result in our dictionary
    auprc_scores[model_name] = auprc_score
    print(f"Calculated AUPRC for {model_name}: {auprc_score:.4f}")

# --- Print the final AUPRC scores, sorted from best to worst ---
print("\n--- Final Area Under the Precision-Recall Curve (AUPRC) Scores ---")
# We sort the dictionary items by their value (the score) in descending order
for model_name, score in sorted(auprc_scores.items(), key=lambda item: item[1], reverse=True):
    # The formatting helps align the scores for easy comparison
    print(f"{model_name:<35}: {score:.4f}")
```

# HYPERPARAMETER TUNING

```python
# We have already trained and evaluated our models.
# Based on our analysis of the AUPRC scores and P-R curves, we select
# the LightGBM model as our champion for further optimization.

# Let's assign the `lgbm_smote` object to a new, more generic variable `best_model`.
# This symbolizes our selection and prepares us for the tuning step.
best_model = lgbm_smote

# Let's print the model to confirm our selection.
print("Selected Best Model for Hyperparameter Tuning:")
print(best_model)
```

```python
# We are defining a dictionary that will hold the hyperparameters we want to tune
# and the range of values we want to test for each one. This is often called
# a "parameter grid" or "parameter distribution".

param_grid = {
    # n_estimators: The number of boosting rounds or trees.
    # We will test a range of values from 100 (our baseline) up to 500.
    # More trees can lead to better performance but also longer training times.
    'n_estimators': [100, 200, 300, 500],
    
    # learning_rate: The step size shrinkage.
    # A smaller learning rate often requires more estimators but can lead to a more robust model.
    # We test values from a relatively high 0.1 down to a very small 0.01.
    'learning_rate': [0.01, 0.05, 0.1],
    
    # num_leaves: The maximum number of leaves in one tree. This is a key parameter for
    # controlling the complexity of the model. A higher value can lead to overfitting.
    # The default is 31. We'll explore values around and above the default.
    'num_leaves': [20, 31, 40, 50],
    
    # max_depth: The maximum depth of a tree. We use -1 for no limit, which is the default,
    # and also test shallower trees to see if they can prevent overfitting.
    # `num_leaves` is usually the more important parameter to tune.
    'max_depth': [-1, 10, 20],
    
    # reg_alpha (L1 regularization): Helps combat overfitting.
    # We'll test a range of values to see if regularization helps.
    'reg_alpha': [0, 0.1, 0.5, 1],
    
    # reg_lambda (L2 regularization): Also helps combat overfitting.
    # Testing a range of values here as well.
    'reg_lambda': [0, 0.1, 0.5, 1]
}

# Let's print the dictionary to see our defined search space.
print("Hyperparameter Grid for Randomized Search:")
# We'll use a nice format to print the dictionary for better readability.
import json
print(json.dumps(param_grid, indent=2))
```

```python
# Import RandomizedSearchCV from scikit-learn's model_selection module
from sklearn.model_selection import RandomizedSearchCV

# --- Step 1: Instantiate the RandomizedSearchCV object ---
# We are creating the 'master engineer' that will conduct our tuning experiment.

random_search = RandomizedSearchCV(
    estimator=best_model,                 # Our champion LightGBM model
    param_distributions=param_grid,       # The hyperparameter grid we defined
    n_iter=25,                            # Number of random combinations to try
    scoring='average_precision',          # The metric to optimize (AUPRC)
    cv=3,                                 # 3-fold cross-validation
    n_jobs=-1,                            # Use all available CPU cores
    random_state=42,                      # For reproducible results
    verbose=1                             # To see progress updates during fitting
)

# --- Step 2: Confirm Instantiation ---
# At this point, no training has been done.
# We have simply created and configured the search object.
# It is now a blueprint for the experiment we are about to run.
print("RandomizedSearchCV object created and configured successfully:")
print(random_search)
```

```python
# Import the time library to measure how long the search takes
import time

# --- Start the Hyperparameter Search ---

# Record the start time
start_time = time.time()

# This is the command that kicks off the entire process described above.
# We are fitting the RandomizedSearchCV object, not the model directly.
# The object will use our balanced training data (X_resampled, y_resampled)
# to conduct its search for the best hyperparameters.
print("Starting the randomized hyperparameter search...")
print(f"The search will test {random_search.n_iter} different parameter combinations.")
print("This may take several minutes to complete...")

random_search.fit(X_resampled, y_resampled)

# Record the end time
end_time = time.time()

# Calculate and print the total duration
duration = end_time - start_time
print(f"\nRandomized search completed in {duration / 60:.2f} minutes.")
```

```python
# The 'random_search' object, having been fitted, now contains the results.

# --- Step 1: Extract the Best Hyperparameters ---
# The .best_params_ attribute returns a dictionary of the winning hyperparameter combination.
best_hyperparameters = random_search.best_params_

# --- Step 2: Extract the Best Score ---
# The .best_score_ attribute returns the mean cross-validated score of the best estimator.
# In our case, this is the best AUPRC score found during the search.
best_cv_score = random_search.best_score_

# --- Step 3: Print the Results for Analysis ---
print("--- Best Hyperparameters Found by Randomized Search ---")

# We import the `json` library to print the dictionary in a nicely formatted way.
import json
print(json.dumps(best_hyperparameters, indent=2))

print(f"\nBest Cross-Validated AUPRC Score from Search: {best_cv_score:.4f}")

# --- Step 4: Compare with the Original Untuned Model ---
# To see if tuning was worthwhile, let's compare this score to the AUPRC of our
# original, untuned LightGBM model from the evaluation step.
# (This assumes you still have the `auprc_scores` dictionary from Step 5 in memory).
original_auprc = auprc_scores['LightGBM (SMOTE)']
print(f"Original Untuned Model's AUPRC on Test Set:     {original_auprc:.4f}")

# Calculate the percentage improvement
improvement = ((best_cv_score - original_auprc) / original_auprc) * 100
print(f"\nPotential Improvement from Tuning: {improvement:.2f}%")
```

```python
# --- SHAP: version-proof, binary LightGBM ---
import shap
import numpy as np

# decide which index is the "positive" class (fraud)
# (LightGBM usually uses [0, 1] = [not fraud, fraud])
pos_idx = 1
if hasattr(final_model, "classes_"):
    # If classes are [0,1] -> pos_idx=1; if [1,0] -> pos_idx=index of 1
    try:
        pos_idx = int(np.where(np.array(final_model.classes_) == 1)[0][0])
    except Exception:
        pos_idx = 1

print("\n--- Creating SHAP explainer (TreeExplainer) ---")
explainer = shap.TreeExplainer(final_model)
print("Explainer created successfully.")

print("\n--- Calculating SHAP values for the test set ---")
print("This may take a few moments to compute...")

def get_pos_class_shap(explainer, X, pos_idx=1):
    """
    Returns SHAP values for the positive class as a 2D array (n_samples, n_features),
    robust to SHAP versions (list-return vs Explanation/ndarray-return).
    Also returns base values (expected value) for the positive class.
    """
    # Preferred new API: call explainer (works in modern shap)
    try:
        explanation = explainer(X)  # shap.Explanation in new versions
        sv = explanation.values            # (n_samples, n_features)
        base_values = explanation.base_values
        # If model has two outputs and SHAP stacked them, sv may still be 2D for binary
        # That's already the positive-class attribution in modern shap for binary models.
        return sv, base_values
    except Exception:
        # Fallback: old API
        sv_raw = explainer.shap_values(X)  # could be list-of-arrays or 2D ndarray
        # expected_value can be scalar or list
        exp_val = getattr(explainer, "expected_value", None)

        if isinstance(sv_raw, list):
            # pick positive class
            sv = sv_raw[pos_idx]
            if isinstance(exp_val, list) or isinstance(exp_val, np.ndarray):
                base_values = np.array(exp_val[pos_idx])
            else:
                base_values = exp_val
        else:
            # already a 2D array = positive class
            sv = sv_raw
            base_values = exp_val
        return sv, base_values

shap_values_pos, base_values_pos = get_pos_class_shap(explainer, X_test, pos_idx=pos_idx)

print("SHAP values calculated successfully.\n")

# --- Sanity prints (match your previous logging style) ---
print(f"Type of shap_values: {type(shap_values_pos)}")
print(f"Shape of our test set (X_test): {X_test.shape}")
print(f"Shape of the SHAP values for the 'Fraud' class: {shap_values_pos.shape}")
print("\nEach row in this SHAP values array corresponds to one transaction in X_test.")
print("Each column corresponds to a feature, quantifying its impact on the fraud prediction.")

# Optional: a quick check for one sample
i = 0
phi_sum = shap_values_pos[i].sum()
# If you use predict_proba, the model output is typically log-odds internally for tree models;
# plotting functions handle this automatically. The base value here corresponds to model output baseline.
print(f"\n[Sample {i}] Sum of SHAP values: {phi_sum:.6f}")
print(f"[Sample {i}] Base value (expected value): {np.array(base_values_pos).ravel()[0]:.6f}")

# Example plots (uncomment if you want visuals)
# shap.summary_plot(shap_values_pos, X_test, plot_type="bar", max_display=20)
# shap.summary_plot(shap_values_pos, X_test)
```
